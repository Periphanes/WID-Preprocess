```python
import csv
import numpy as np
import pandas as pd
import os

from pathlib import Path
import chardet
```

```python
import matplotlib.pyplot as plt
```

```python
directory = 'C:\MAU_dataset\data'
data = {}
meta = {}

for filename in os.scandir(directory):
    if filename.is_file():
        f = filename.path
        if f[-3:] == "csv":
            print(f)
            if f.split("\\")[-1][4:7] == "cou":
                countries = pd.read_csv(f, delimiter=";")
            elif f.split("\\")[-1][4:7] == "dat":
                country_code = f.split(".")[-2].split("_")[-1]
                data[country_code] = pd.read_csv(f, delimiter=";")
            elif f.split("\\")[-1][4:7] == "met":
                detected = chardet.detect(Path(f).read_bytes())
                encoding = detected.get("encoding")
                assert encoding, "Unable to detect encoding, is it a binary file?"
                
                country_code = f.split(".")[-2].split("_")[-1]
                meta[country_code] = pd.read_csv(f, delimiter=";", encoding=encoding)
```

```python
for index, row in countries.iterrows():
    print(str(row["alpha2"]), end="/")
    if str(row["alpha2"]) == "nan":
        countries.drop(index, inplace=True)
```

```python
delete_countries = ["nan", "XC" , "XE", "XK", "XQ" ,"XQ-MER", "Al", "Cl", "Cl "]
for index, row in countries.iterrows():
    if str(row["alpha2"]) in delete_countries:
        countries.drop(index, inplace=True)
```

```python
# The variables in metadata have different names to the ones in the real data
# For data variable names: {ORG_NAME}{000 - Number}{Letter}
# For metadata, this is changed to -> {ORG_NAME}{Letter}{000 - Number}

def dat_to_met_variable(varname):
    return varname[:-4] + varname[-1] + varname[-4:-1]
def met_to_dat_variable(varname):
    return varname[:-4] + varname[-3:] + varname[-4]
```
