```python
import csv
import numpy as np
import pandas as pd
import os

from pathlib import Path
import chardet
```

```python
directory = 'C:\MAU_dataset\data'
data = {}
meta = {}

for filename in os.scandir(directory):
    if filename.is_file():
        f = filename.path
        if f[-3:] == "csv":
            print(f)
            if f.split("\\")[-1][4:7] == "cou":
                countries = pd.read_csv(f, delimiter=";")
            elif f.split("\\")[-1][4:7] == "dat":
                country_code = f.split(".")[-2].split("_")[-1]
                data[country_code] = pd.read_csv(f, delimiter=";")
            elif f.split("\\")[-1][4:7] == "met":
                detected = chardet.detect(Path(f).read_bytes())
                encoding = detected.get("encoding")
                assert encoding, "Unable to detect encoding, is it a binary file?"
                
                country_code = f.split(".")[-2].split("_")[-1]
                meta[country_code] = pd.read_csv(f, delimiter=";", encoding=encoding)
```
