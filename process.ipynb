```python
# For Normal Run, Start Here

import pickle
pickle_dir = 'C:\MAU_dataset\pickle\ '

with open(pickle_dir.strip() + 'countries.pickle', 'rb') as handle:
    countries = pickle.load(handle)

with open(pickle_dir.strip() + 'data.pickle', 'rb') as handle:
    data = pickle.load(handle)

with open(pickle_dir.strip() + 'meta.pickle', 'rb') as handle:
    meta = pickle.load(handle)
```

```python
# The variables in metadata have different names to the ones in the real data
# For data variable names: {ORG_NAME}{000 - Number}{Letter}
# For metadata, this is changed to -> {ORG_NAME}{Letter}{000 - Number}

def dat_to_met_variable(varname):
    return varname[:-4] + varname[-1] + varname[-4:-1]
def met_to_dat_variable(varname):
    return varname[:-4] + varname[-3:] + varname[-4]
```

```python
all_vars = []
for meta_row_index, meta_row in meta["KR"].iterrows():
    var_name = met_to_dat_variable(meta_row.variable)
    found_all = True
    
#     print("Variable Name: ", var_name)
    
    for year in range(1980, 2015):
        quer = data["KR"].query('variable == "' + var_name + '"').query('year == ' + str(year))
        if quer.empty:
            found_all = False
#         else:
#             print(year, end="/")
#     print("")
    if found_all:
        print(var_name, end=" / ")
        all_vars.append(var_name)
```
